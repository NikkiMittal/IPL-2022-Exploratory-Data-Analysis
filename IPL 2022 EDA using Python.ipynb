# IPL 2022 Exploratory Data Analysis using Python

```python
#Importing required libraries:

import pandas as pd                
import seaborn as sns              
import matplotlib.pyplot as plt    
%matplotlib inline  

   
```

```python
#Reading data:

data = pd.read_csv('IPL2022.csv')
print(data.shape)                  #printing the number of rows and columns respectively
data.head(5)                       #printing the top 5 rows
```

Here, we find that there are 74 entries in total for 20 columns. The dataset contains all the information about the teams, stadiums, top scorers etc.

```python
#Understanding the data more:

data.describe()    #for statistical description of data
```

```python
#To check the need of data cleaning:

data.info()      #gathering information about the non null count and data type
```

```python
data.isnull().sum()     #checking if there is any null value
```

Here, we can clearly see that the dataset does not have any null or missing values. The dataset is cleaned.

### Visualizing the data

```python
# Analyzing the number of matches won by teams in IPL 2022:

plt.figure(figsize=(10,5))
a = sns.countplot(x='match_winner', data = data)
a.set_xticklabels(a.get_xticklabels(), rotation=10)
a.set_xlabel("Match Winner", fontsize = 15)
a.set_ylabel("Count", fontsize = 15)
a.set_title('Number of matches won in IPL 2022', fontsize = 20)
```

It is found that in IPL 2022, the maximum number of matches have been won by Gujarat followed by Rajasthan. Gujarat led the tournament.

```python
# To analyse the player of the matches data:

plt.figure(figsize=(20,5))
b = sns.countplot(x='player_of_the_match', data = data)
b.set_xticklabels(b.get_xticklabels(), rotation=90)
b.set_xlabel("Player of the match", fontsize = 15)
b.set_ylabel("Count", fontsize = 15)
b.set_title('Player of the matches in IPL 2022', fontsize = 20)
```

It is found that Kuldeep Yadav has shown an exemplary performance by becoming the player of the match in 4 matches, followed by Jos Buttler.

```python
#Analyzing the best bowling players in IPL 2022:

plt.figure(figsize=(20,5))
c = sns.countplot(x='best_bowling', data = data)
c.set_xticklabels(c.get_xticklabels(), rotation=90)
c.set_xlabel("Best Bowling player", fontsize = 15)
c.set_ylabel("Count", fontsize = 15)
c.set_title('Best bowling players in IPL 2022', fontsize = 20)
```

Clearly, it is found that Yuzvendra Chahal has emerged as the best bowling player in maximum number of matches, followed by Rashid Khan.

```python
# Analyzing the top scorers of the matches in IPL 2022:

plt.figure(figsize=(20,5))
d = sns.countplot(x='top_scorer', data = data)
d.set_xticklabels(d.get_xticklabels(), rotation=90)
d.set_xlabel("Top scorer", fontsize = 15)
d.set_ylabel("Count", fontsize = 15)
d.set_title('Top scorers in IPL 2022', fontsize = 20)
```

Jos Buttler has become the top scorer in 7 matches that he played while Quinton de Kock becomes the top scorer in 5 matches. Both these players have shown outstanding performances.

```python
#Distribution of the high scores:

data['highscore'].plot.hist()
plt.title('Distribution of the high scores')
plt.xlabel('High Scores')
plt.ylabel('Frequency')
plt.show()
```

The plot shows the distribution of the high scores. Mostly top scorers have got high score in the range of 50 to 70. 

```python
# To check the stadiums in which maximum and minimum number of matches were played:

plt.figure(figsize=(20,5))
f = sns.countplot(x='venue', data = data)
f.set_xticklabels(f.get_xticklabels(), rotation=10)
f.set_xlabel("Venue", fontsize = 18)
f.set_ylabel("Count", fontsize = 18)
f.set_title('Number of matches played at different venues in IPL 2022', fontsize = 20)
```

In IPL 2022, the maximum number of matches have been played in Wankhede Stadium Mumbai, while minimum number of matches have been played in Eden Gardens, Kolkata and Narendra Modi Stadium, Ahmedabad.
